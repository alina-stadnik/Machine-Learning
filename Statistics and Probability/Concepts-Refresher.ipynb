### **Using Python to calculate mean, median and mode:**

```python
def mean(list):
    """"
    Function to calculate the mean
    Args:
        It receives a list of values 
    Returns:
        It returns 0 if it's an empty list
        It returns the mean for a list of values
    """

    if len(list)==0: #check if length is zero
        return 0
    else:
        sum=0 #initialize the sum
        for i in list:
            sum+=i #total sum
            mean=sum/len(list) #average
        return mean
```

```python
mean([5,6,7,8,9])
```

```python
def median(list):
    """"
    Function to calculate the median
    Args:
        It receives a list of values 
    Returns:
        It returns 0 if it's an empty list
        It returns the median for a list of values
    """
    if len(list)==0:  #check if length is zero
        return 0
    else:
        list.sort() #order in ascending order
        midpoint=len(list)//2 #to find middle term
        count=len(list) #to find length
        if count%2==1: #the length is odd
            median=list[midpoint]
        else: 
            median=list[midpoint]+list[midpoint-1]/2 #incase of even terms
    return median
```

```python
median([5,6,7,8,9])
```

```python
median([1,0.2,4])
```

```python
def mode(list):
    """"
    Function to find the mode
    Args:
        It receives a list of values 
    Returns:
        It returns the mode for a list of values
    """
    list.sort()
    list_one=[]
    i = 0
    while i < len(list) : 
        list_one.append(list.count(list[i])) 
        i += 1 
        # count how many times each element appears in the list
        d1 = dict(zip(list, list_one)) 
        # using a dict: pair each element from the sorted list with its corresponding count
        new_list={k for (k,v) in d1.items() if v == max(list_one) } # find the keys with max count in dict 
    return new_list
```

```python
mode([2,5,8,8,8,9,8,8,7])
```

```python
mode([2,5,8,8,8,9,8,8,7,7,7,7,7])
```

```python
from math import sqrt

def std(list):
    """"
    Function to find the standard deviation
    Args:
        It receives a list of values 
    Returns:
        It returns 0 if it's an empty list
        It returns the std of the list
    """
    if len(list)==0: #check if length is zero
        return 0
    
    std_deviation = sqrt(sum([(number - mean(list)) ** 2 for number in list]) / (len(list) - 1)) #mathematical formulation
    return std_deviation
```

**Poisson Probability Mass Function**

Very specific application: it looks like a normal distribution, but it's different. The idea is if you have **some information about the average number of things that happen in some given time period**, this can give you a way to predict the odds of getting some other value instead on a given future day.

> Example: My website gets on average 500 visits per day. What's the odds of getting 550?

```python
from scipy.stats import poisson

mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```

**Moments**

- Quantitative measures of the shape of a probability density function.

- The first moment is the mean;
- The second moment is the variance;
- The third moment is 'skew' 
    - how 'lopsided'is the distribution?
    - a distribution with a longer tail on the left will be skewed left, and have a negative skew.

![image](https://storage.googleapis.com/tb-img/production/20/11/F1_Priya_Shraddha_24.11.2020_D4.png)

- The fourth moment is 'kurtosis'
    - how thick is the tail, and how sharp is the peak, compared to a normal distribution?

![img](https://ars.els-cdn.com/content/image/3-s2.0-B9780323917483000045-f03-07-9780323917483.jpg)

```python
vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
# The first moment is the mean 
# this data should average out to about 0:
np.mean(vals)
```

```python
# The second moment is the variance:
np.var(vals)
```

```python
# The third moment is skew 
# this data is nicely centered around 0, it should be almost 0
import scipy.stats as sp

sp.skew(vals)
```

```python
# The fourth moment is "kurtosis", which describes the shape of the tail
# For a normal distribution, this is 0
sp.kurtosis(vals)
```

- Box and Whisker Plot

Useful for visualizing the spread and skew of data.

The red line represents the median of the data, and the box represents the bounds of the 1st and 3rd quartiles.

So, half of the data exists within the box.

The dotted-line "whiskers" indicate the range of the data - except for outliers, which are plotted outside the whiskers. Outliers are 1.5X or more the interquartile range.

> This example below creates uniformly distributed random numbers between -40 and 60, plus a few outliers above 100 and below -100:

```python
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
plt.boxplot(data)
plt.show()
```

- Heatmap

A heat map allows to plot tabular, 2D data of some sort, with colors representing the individual values in each cell of the 2D table.

In this example, it'll be created a pivot table from the original dataframe, to create a 2D table that contains the average MPG ratings for every combination of number of cylinders and engine displacement.

The resulting heatmap shows all of the engine displacement values along the X axis, and all of the cylinder values along the Y axis. For each cell of the table, the actual average MPG rating for that combination of cylinders and engine displacement is represented not as a number, but as a color that ranges from dark for small values, and light for larger values.

And, this allow to visualize a clear trend where things get progressively darker as it moves from the top-left of the graph to the bottom-right. This makes sense: **higher MPG ratings are associated with lower numbers of cylinders, and lower engine displacment values**. By the time it gets to an 8-liter 16-cylinder engine, the average MPG is at its worst of about 12, represented by the color black.

This particular graph has a lot of missing data, but the heatmap deals with that in a satisfactory way. A 3-cylinder 8-liter engine simply does not exist.

```python
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean')
sns.heatmap(df2)
```

### **Covariance and Correlation**

**Covariance**:
- It is the measure of how much two random variables change together. Indicates the direction of the linear relationship between two variables.
- Covariance can take on any value, and its sign indicates the direction of the relationship:
    - Positive: Indicates a positive relationship (as one variable increases, the other tends to increase).
    - Negative: Indicates a negative relationship (as one variable increases, the other tends to decrease).

**Correlation**:
- It is a standardized measure of the strength and direction of the linear relationship between two variables.
- The correlation coefficient ranges from -1 to 1:
    - **1**: indicates a perfect positive linear relationship.
    - **-1**: indicates a perfect negative linear relationship.
    - **0**: indicates no linear relationship.
- Correlation is a more interpretable measure because it is standardized and allows for easy comparison of the strength and direction of relationships between different pairs of variables.

In summary, while covariance measures the direction of the linear relationship between two variables, correlation provides a standardized measure that facilitates comparison and interpretation of the strength and direction of this relationship.

Example: let's say we work for an e-commerce company, and they are interested in finding a correlation between page speed (how fast each web page renders for a customer) and how much a customer spends.

```python
from pylab import *

def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

Now we'll make our fabricated purchase amounts an actual function of page speed, making a very real correlation. 

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

> **The negative value indicates an inverse relationship; pages that render in less time result in more money spent.**

What does this value mean? 

Covariance is sensitive to the units used in the variables, which makes it difficult to interpret. Correlation normalizes everything by their standard deviations, giving it an easier to understand value that ranges from -1 (for a perfect inverse correlation) to 1 (for a perfect positive correlation):

```python
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y) / stddevx / stddevy  

correlation(pageSpeeds, purchaseAmount)
```

Using numpy with numpy.corrcoef: it returns a matrix of the correlation coefficients between every combination of the arrays passed in:

```python
np.corrcoef(pageSpeeds, purchaseAmount)
```

(It doesn't match exactly just due to the math precision available on a computer)

It is possible to force a perfect correlation by fabricating a totally linear relationship (again, it's not exactly -1 just due to precision errors, but it's close enough to tell if there is a really good correlation here):

```python
purchaseAmount = 100 - pageSpeeds * 3

scatter(pageSpeeds, purchaseAmount)

correlation (pageSpeeds, purchaseAmount)
```

> **Correlation does not imply causality!**
